### P-values pequeños significan que hay evidencia más fuerte a favor de la hipótesis alternativa

Probar la hipótesis nula de que la asimetría de la población de la que se extrajo la muestra es la misma que la de una distribución normal:

Se acepta la Hipotesis Alternativa, por lo cual, no es conveniente utilizar la correlación de pearson.

```python
from scipy.stats import skewtest, levene, shapiro

skw = skewtest(df_1)
plt.figure(figsize=(14,4))
plt.plot(skw.pvalue, lw=3)
plt.xticks(range(len(df_1.columns)),df_1.columns, rotation=90)
plt.ylabel("p-value")
plt.xlabel("Feature")
plt.grid()
plt.legend('Score')
plt.title("Skewtest")
```

### Al analizar la matriz de correlación presentada, se puede observar que:

La variable numero_de_fracasos tiene una correlación negativa fuerte (-0.556859) con la variable estado_final, lo que sugiere que hay una relación inversa entre estas dos variables.

La variable antecedente_fracaso_a. no tiene una correlación positiva fuerte (0.556931) con la variable estado_final, lo que sugiere que hay una relación directa entre estas dos variables.

La variable antecedente_fracaso_b. si tiene una correlación negativa fuerte (-0.556931) con la variable estado_final, lo que sugiere que hay una relación inversa entre estas dos variables.

En general, la matriz de correlación sugiere que las variables numero_de_fracasos, antecedente_fracaso_a. no y antecedente_fracaso_b. si pueden estar relacionadas con el estado_final de los pacientes, mientras que la persistencia y el tiempo_en_tratamiento_arv pueden no tener una relación significativa con el estado_final. Sin embargo, es importante tener en cuenta que el análisis de correlación no implica causalidad y que se necesitan más análisis para llegar a conclusiones definitivas.

```python
plt.figure(figsize=(16,8))
cor_matrix = df_1.corr('kendall')
cor_matrix_filter = cor_matrix[(cor_matrix > 0.5) | (cor_matrix < -0.5)]
cor_matrix_filter = cor_matrix_filter
np.fill_diagonal(cor_matrix_filter.values, np.NaN)
sns.heatmap(cor_matrix_filter ,vmin=-1, vmax=1,cmap="jet", center=0, annot=True, fmt=".0%", square=True)
```

No necesariamente debes escoger la variable con la correlación más alta para explicar el estado final de éxito en el tratamiento. Es importante tener en cuenta que la correlación entre dos variables no implica causalidad, es decir, una alta correlación entre una variable y el estado final de éxito no significa necesariamente que esa variable sea la causa del éxito.

Además, el éxito en el tratamiento de VIH es un fenómeno complejo que puede estar influenciado por múltiples factores y no se puede explicar completamente por una sola variable.
